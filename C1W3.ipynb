```python
import matplotlib.pyplot as plt
import numpy as np

def plot_vectors(list_v, list_label, list_color):
    # Create a new figure with specified size
    _, ax = plt.subplots(figsize=(10, 10))
    
    # Set the font size for tick labels
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
    
    # Set the tick marks
    ax.set_xticks(np.arange(-10, 10))
    ax.set_yticks(np.arange(-10, 10))
    
    # Set the plot limits
    plt.axis([-10, 10, -10, 10])
    
    # Plot each vector
    for i, v in enumerate(list_v):
        # Ensure v is a 2D array
        v = np.array(v).reshape(-1, 1)
        
        # Calculate the sign array for text positioning
        sgn = 0.4 * np.array([[1] if j==0 else [j] for j in np.sign(v)])
        
        # Plot the vector
        plt.quiver(0, 0, float(v[0]), float(v[1]), 
                  color=list_color[i], angles='xy', 
                  scale_units='xy', scale=1)
        
        # Add the label
        ax.text(float(v[0])-0.2+float(sgn[0]), 
                float(v[1])-0.2+float(sgn[1]), 
                list_label[i], fontsize=14, 
                color=list_color[i])

    # Add grid and set aspect ratio
    plt.grid()
    plt.gca().set_aspect("equal")
    plt.show()

# Example usage
v = np.array([[1], [3]])  # Define a vector
plot_vectors([v], ["$v$"], ["black"])
```
